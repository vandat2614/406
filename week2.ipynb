## **Open CV**

```python
img_bgr = cv2.imread("goku.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

print(f'Type: {type(img_rgb)}')
print(f'Shape: {img_rgb.shape}')
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img_bgr)
plt.axis('off')
plt.title("BGR")

plt.subplot(1, 2, 2)
plt.imshow(img_rgb)
plt.axis('off')
plt.title("RGB")
```

```python
gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
cv2.imwrite("goku_gray_cv2.png", gray)
```

```python
img = cv2.imread("goku.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

red, green, blue = cv2.split(img)
# red, green, blue = img[:,:,0], img[:,:,1], img[:,:,2]


plt.subplot(1, 3, 1)
plt.imshow(red, cmap='gray')
plt.title("Red")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(green, cmap='gray')
plt.title("Green")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(blue, cmap='gray')
plt.title("Blue")
plt.axis('off')
```

```python
img = cv2.imread("goku.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_copy_raw = img

img_copy_raw[:] = 0

plt.subplot(1, 2, 1)
plt.imshow(img_copy_raw)
plt.title("Raw")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img)
plt.title("Original")
plt.axis('off')
```

## **Manipulating**

```python
img = Image.open("goku.jpg")
img_flip = ImageOps.flip(img) # Lật ảnh theo chiều dọc
img_mirror = ImageOps.mirror(img) # Lật ảnh theo chiều ngang
img_rotate = img.rotate(120) # Xoay ảnh 90 đ

plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_flip)
plt.title("Flip")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_mirror)
plt.title("Mirror")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_rotate)
plt.title("Rotate")
plt.axis('off')
```

```python
img = cv2.imread("goku.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_flip = cv2.flip(img, 0)
img_mirror = cv2.flip(img, 1)
img_rotate = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_flip)
plt.title("Flip")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_mirror)
plt.title("Mirror")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_rotate)
plt.title("Rotate")
plt.axis('off')
```

# **Crop**

```python
img = cv2.imread("goku.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

upper, lower, left, right = 400, 670, 200, 400
cropped = img[upper:lower, left:right]

plt.imshow(cropped)
plt.axis('off')
```

```python
img = Image.open("goku.jpg")

upper, lower, left, right = 400, 670, 200, 400
img_crop = img.crop((left, upper, right, lower))

plt.imshow(img_crop)
plt.axis('off')
```

## **Draw**

```python
img = Image.open("goku.jpg")

img_draw = ImageDraw.Draw(img)

# left, upper, right, lower
img_draw.rectangle(xy=[200, 400, 400, 670], outline='red', width=15)
img_draw.rectangle(xy=[500, 600, 600, 870], fill='blue')
img_draw.text(xy=(200,200),text="AAAAAAAAAAAAA"*20,fill=(0,0,0))

plt.imshow(img)
plt.axis('off')
```

```python
img.paste(img_crop, box=(700, 600)) # left, upper

plt.imshow(img)
plt.axis('off')
```

```python
img = cv2.imread("goku.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2.rectangle(img, pt1=(200, 400), pt2=(400, 670), color=(255, 0, 0), thickness=15)
cv2.rectangle(img, pt1=(500, 600), pt2=(600, 870), color=(0, 0, 255), thickness=-1) # -1 for fill
# cv2.putText(img, text="Goku", org=(200, 200), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=10, color=(0, 255, 0), thickness=10)
cv2.putText(img, text="Goku", org=(200, 200), fontFace=1, fontScale=10, color=(0, 255, 0), thickness=10)


plt.imshow(img)
plt.axis('off')
```
