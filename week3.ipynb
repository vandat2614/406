```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## **Histogram**

```python
img = cv2.imread('goku.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (128, 128))

# danh sách ảnh, danh sách kênh, mask, số bin, range của giá trị
colors = ('b', 'g', 'r')
for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)

plt.title('Color Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

## **Intensity Transformations**

```python
img = cv2.imread('goku.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_copy = img.copy()
img_copy = 2 * img_copy + 1

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_copy)
plt.title('Brightened Image')
plt.axis('off')
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_copy = cv2.calcHist([img_copy], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(hist, color='b')

plt.subplot(1, 2, 2)
plt.plot(hist_copy, color='r')
```

## **Brightness and Contrast**

```python
alpha = 1 # Contrast control
beta = 100 # Brightness control

img = cv2.imread('goku.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
new_image = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(new_image)
plt.title('Adjusted Image')
plt.axis('off')
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_new = cv2.calcHist([new_image], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(hist, color='b')

plt.subplot(1, 2, 2)
plt.plot(hist_new, color='r')
```

## **Histogram Equalization**

```python
img = cv2.imread('goku.png', cv2.IMREAD_GRAYSCALE) 
img_copy = cv2.equalizeHist(img_copy) # chỉ cân bằng cho ảnh xám

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_copy, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_copy = cv2.calcHist([img_copy], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(hist, color='b')

plt.subplot(1, 2, 2)
plt.plot(hist_copy, color='r')
```

## **Translation**

```python
img = cv2.imread('goku.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

tx, ty = 400, 100
height, width, _ = img.shape
M = np.float32([[1, 0, tx], [0, 1, ty]])

translated_img = cv2.warpAffine(img, M, dsize=(width, height))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(translated_img)
plt.title('Translated Image')
plt.axis('off')
```

## **Rotation**

```python
theta = 60 # ngược hướng kim đồng hồ
height, width, _ = img.shape

M = cv2.getRotationMatrix2D((width//2, height//2), angle=theta, scale=2)
rotated_img = cv2.warpAffine(img, M, dsize=(width, height))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_img)
plt.title('Rotated Image')
plt.axis('off')
```
